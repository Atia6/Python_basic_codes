Create an array containing 1000 random floats between 1 to 10.<br>
Then show the histogram of the generated data.<br>


Integar data code

```python
import random
import matplotlib.pyplot as plt

random_numbers = [random.randint(1, 10) for _ in range(1000)]

plt.hist(random_numbers, bins=10, edgecolor='black')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')
#plt.xticks(range(1, 11))
plt.show()
```

For float numbers

```python
import numpy as np
import matplotlib.pyplot as plt

random_floats = np.random.uniform(1, 10, size=1000)

plt.hist(random_floats, bins=10, edgecolor='black')
plt.xlabel('Floats')
plt.ylabel('Frequency')
plt.title('Histogram of Random Floats')
plt.show()
```
